```python
# Download historical stock data using yfinance and save it to a CSV file.
import yfinance as yf
import os

def download_stock_data(ticker, start_date, end_date):
    # Create the output directory if it doesn't exist
    os.makedirs('./stock_data', exist_ok=True)

    # Download historical stock data
    stock_data = yf.download(ticker, start=start_date, end=end_date)
    
    # Save the data to a CSV file
    output_file = os.path.join('./stock_data', f"{ticker}_historical_data.csv")
    stock_data.to_csv(output_file)
    print(f"Data for {ticker} saved to {output_file}")


if __name__ == "__main__":   

    ticker_list = ["AAPL", # Apple Inc. (Tech company)
                   "MSFT", # Microsoft Corporation (Tech company)
                   "NVDA", # NVIDIA Corporation (Tech company)
                   "GOOGL", # Alphabet Inc. (Google) (Communication / media)
                   "AMZN", # Amazon.com, Inc. (Consumer discretionary)
                   "WMT", # Walmart Inc. (Retail)
                   "META", # Meta Platforms, Inc. (Social media / communication)
                   "TSLA", # Tesla, Inc. (Automotive / energy)
                   "JPM", # JPMorgan Chase & Co. (Financial services)
                   "GS"    # Goldman Sachs Group, Inc. (Financial services)
                   ]  
    
    start_date = "2024-01-01"
    end_date = "2026-01-01"

    for ticker in ticker_list:
        download_stock_data(ticker, start_date, end_date)
```
